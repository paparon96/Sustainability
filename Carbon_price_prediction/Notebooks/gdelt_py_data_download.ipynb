# Import packages

```python
import gdelt
import pandas as pd
```

# Parameters

```python
start_date = '2022-07-25'
end_date = '2022-08-01'
# dates = pd.date_range('2019-01-01', '2019-01-15', freq='b')
# dates = [date.strftime("%Y-%m-%d") for date in dates]
dates = [start_date, end_date]

actor_filters = ["THE EUROPEAN UNION", "EUROPEAN UNION"]
```

# Data import

```python
gd = gdelt.gdelt()

events = gd.Search(dates, table='events', output='gpd', normcols=True, coverage=False)
```

# Data preprocessing

```python
filtered_events = events[events.actor1name.isin(actor_filters)]
```
