# Import packages

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

# Constants / Parameters

```python
n = 10
r = 0.1
k = r / 365

mean = 0
sigma = 0.1
```

# Data import (or dummy data generating process)

```python
returns = np.random.normal(mean, sigma, size=n)
predictions = np.random.normal(mean, sigma, size=n)

rf_strat_returns = np.ones(n) * k # risk-free strategy returns
```

# Strategy simulation

```python
# Simulate active strategy returns
active_strat_returns = rf_strat_returns.copy()
active_strat_returns[predictions > 0] = returns[predictions > 0]
```

```python
# Calculate cumulative returns of the strategies
base_value = 100
active_strat_cumulative_returns = base_value * pd.Series(1 + active_strat_returns).cumprod()
rf_strat_cumulative_returns = base_value * pd.Series(1 + rf_strat_returns).cumprod()
```

# Visualize results

```python
plt.plot(active_strat_cumulative_returns, label="Active strategy")
plt.plot(rf_strat_cumulative_returns, label="Risk-free strategy")
plt.xlabel("Time (days)")
plt.ylabel("Rebased portfolio levels")
plt.legend()
plt.show()
```
