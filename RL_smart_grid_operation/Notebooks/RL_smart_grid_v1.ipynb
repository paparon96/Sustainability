# Import packages

```python
import grid2op

import matplotlib.pyplot as plt
import numpy as np
```

# Parameters

```python
num_train_iter = 100
num_eval_iter = 10

base_save_path = "./trained_models"
```

# DoubleDuelingDQN

## Model training

```python
from l2rpn_baselines.DoubleDuelingDQN import train
from l2rpn_baselines.DoubleDuelingDQN import evaluate

env = grid2op.make()
res = train(env, save_path=base_save_path, iterations=num_train_iter)
```

## Model evaluation

```python
env = grid2op.make()
res = evaluate(env, load_path=base_save_path + "/DoubleDuelingDQN.h5", nb_episode=num_eval_iter)
```

```python
type(res)
```

```python
DoubleDuelingDQN_rewards = [i[2] for i in res]
```
